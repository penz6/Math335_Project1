### Abstract

We predicted that we would be able to predict the average housing price per month for a zip code based on factors including location, house inventory, and other housing factors. We gave the model data from 2012-2023, and predicted data from 2023-present.

### Data Exploration

In our dataset we had over 6,000,000 data points from month to month in different zip codes. To determine our input factors, we decided to include data that would reasonably be available to the model when predicting future prices. We narrowed it down to 13 data points, including location-based features such as state, city, and region, property details such as property type, market activity measures like inventory and homes sold, timing indicators like year and seasonal month patterns, and past pricing information through lagged sale prices, all of which were used to help the model predict median sale price more accurately. We created some of our own inputs to the model, we gave it the median sale price three months before in the same region, as well as scaling inputs and encoding data into numerical format.

### Model Creation

We created models with two different amounts of layers, single layer models, and double layer models. We ran 10, 25, 50, 75, 100, 250, and 500 epochs for each model. For the single layer model we ran 10, 25, 50, 75, 100, 125, 200, and 350 neurons. For the 2 layer model we ran (10, 5), (25, 10), (50, 25), (75, 50), (100, 75), (125, 100), (200, 125), and (350, 200) neurons. We chose these options as it was a wide variety of options, and since the dataset was so large, it would not have fit within our time frame for the project. For our split of test data and train data, we split by date rather than size. We ended up with 1,158,819 training rows, 296,235 validation rows, and 577,305 test rows.

### Hyperparameter Tuning

We tuned our hyperparameter manually using a grid search by running multiple fixed model sizes and epoch values, and picking the best one based on validation RMSE score. We fixed the learning rate at 0.0001, and a batch size of 4098. We did this because of the size of the data and the lack of time to test many batch sizes and learning rates, and after doing some research we determined that these values were average for our dataset size and should give reasonable performance.

### Model Results

Our best performing model was the single layer neuron model with 125 neurons and 500 epochs. We got a validation RMSE of 197,080 and a test RMSE of 273,850. Our best performing two layer model was the 100 neuron first layer and 75 neuron second layer variant at 500 epochs. It had a validation RMSE of 197,617 and a test RMSE of 281,345.

### Statistical Analysis

The single layer model performed better than the
multi layer model on the shared test set. It had a
lower RMSE of $273,850.07 compared to
$281,345.02, a slightly stronger correlation with
the actual values at r = 0.8363, and a significantly
lower absolute error based on the paired t-test (p <
1e-12). Both models failed the normality test, so
their residuals were not normally distributed, but
overall the single layer model was the strongest
